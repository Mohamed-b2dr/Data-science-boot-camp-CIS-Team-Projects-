```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
theta0 = 200
theta1 = 90
```

```python
xs =[i for i in range(15)]
```

```python
ys = [theta0 + theta1*x for x in xs]
```

```python
Ecommerce = pd.read_csv('Ecommerce Customers')
```

```python
sns.scatterplot( data = Ecommerce, x = 'Length of Membership', 
                y = 'Yearly Amount Spent')
sns.lineplot(x=xs, y=ys)
plt.show()
```
